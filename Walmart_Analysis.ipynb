```python
df['total_revenue'] =  df.unit_price * df.quantity
```

```python
import pymysql
from sqlalchemy import create_engine
```

```python
engine = create_engine("mysql+pymysql://root:sujal@localhost/walmart")
```

```python
df.to_sql(name = 'walmart' , con = engine, if_exists = 'append')
```

```python
pd.read_sql("select * from walmart", engine)
```

```python
# Find different payment method and number of transaction , number of quantity sold
pd.read_sql('''select payment_method, count(*) as number_of_transaction, sum(quantity) as total_quantity from walmart
                group by payment_method
                order by total_quantity desc''', engine)
```

```python
# Display the highest rated category in each branch , displaying the branch, category, avg rating
pd.read_sql('''
select * from (
select 
branch,
category, 
avg(rating) as avg_rating,
rank() over(partition by branch order by avg(rating)desc) as ranked
from walmart 
group by branch, category
) as ranks
where ranked = 1
''', engine)
```

```python
# Idemtify the busiest day for each branch based on the number of transaction
pd.read_sql('''
select * from (
SELECT branch,
DAYNAME(datetime) AS day,
count(*) as no_transactions,
rank() over(partition by branch order by count(*) desc) as ranked
FROM walmart
group by branch, day
order by branch, no_transactions desc
) as ranks
where ranked = 1
''', engine)
```

```python
# Calculate the total quantity of items sold per payment method list payment method and total_quantity
pd.read_sql('''
select payment_method,
sum(quantity) as total_quantity
from walmart
group by payment_method
''', engine)
```

```python
# Determine the average, minimum, and maximum rating of products for each city 
# Display the city, average_rating, min_rating and max_rating

pd.read_sql('''
select City,
category,
avg(rating),
min(rating),
max(rating) 
from walmart
group by city,category
''', engine)
```

```python
# Calculate the total profit for each category by considering 
# total profit as unit_price * quantity * profit_margin.
#List category and total profit ordered from desc
pd.read_sql('''
select category,
sum(total_revenue * profit_margin) as total
from walmart
group by category
order by total desc
''', engine)
```

```python
# Determine the most common payment method for each branch Display branch and the preferred payment method
pd.read_sql('''
select * from (
seect
branch,
payment_method,
count(*) as pop,
rank() over(partition by branch order by count(*) desc) as ranked
from walmart
group by branch, payment_method
order by branch
) as ranks
where ranked = 1
''', engine)
```
